```python
scores = np.random.randint(0, 100, 100)
```

```python
bins = [0, 20, 40, 60, 80, 100]
plt.hist(scores, color="skyblue", edgecolor = "black", bins = 8)
plt.xlabel("Marks of students")
plt.ylabel("Count")
plt.axvline(x = 33, color = "blue", label = 'passing marks')
plt.title("Distribution of marks of students")
```

```python
#plotting box plots for multiple datasets
np.random.seed(0)
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 15, 100)

# plt.boxplot([group1, group2], tick_labels = ["Group 1", "Group 2"], showmeans = True)
#to plot horizontal boxplot
# plt.boxplot([group1, group2], vert = False, tick_labels = ["Group 1", "Group 2"])
#to show mean
# plt.boxplot([group1, group2], vert = False, tick_labels = ["Group 1", "Group 2"], showmeans = True)
#to annotate the shown mean values
plt.text(1.1, np.mean(group1), f"Mean : {np.mean(group1):.2f}", color = "red")
plt.text(2.1, np.mean(group2), f"Mean : {np.mean(group2):.2f}", color = "red")
# to change whisker length
plt.boxplot([group1, group2], vert = True, tick_labels = ["Group 1", "Group 2"], showmeans = True, whis=2.0)
plt.grid()
```

```python
#Creating Stack plots

days = ["Sun", "Mon", "Tues", "Wed", "Thurs", "Fri", "Sat"]
direct = [x for x in range(50, 120, 10)]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]

# pie chart to show data distribution on monday
# plt.pie(
#     [direct[0], organic[0], social[0]],
#     labels = ["Direct", "Organic", "Social"],
#     wedgeprops={
#         "edgecolor" : "black"
#     },
#     autopct = "%1.2f%%"
# )

# plt.legend()


plt.stackplot(days, [direct, organic, social], labels = ["Direct", "Organic", "Social"])
plt.title("Marketing data for the week")
plt.xlabel("Days of weeks")
plt.ylabel("No. of customers")
plt.legend()
```
