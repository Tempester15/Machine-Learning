```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
def generate_data():
  x = np.linspace(0, 2*np.pi, 200)
  y = np.sin(x)
  z = np.cos(x)
  return x, y, z


def plot_line_graph(ax, data1, data2, title, xlabel, ylabel):
  axs[ax].plot(data1, data2, 'g-')
  axs[ax].set_title(f"{title}")
  axs[ax].set_xlabel(f"{xlabel}")
  axs[ax].set_ylabel(f"{ylabel}")




x_axis_data, y_axis_data, z_axis_data = generate_data()

fig, axs = plt.subplots(1, 2, figsize = (10, 2))

plot_line_graph(0, x_axis_data, y_axis_data, "Sine of X graph", "X", "Sin(x)")
plot_line_graph(1, x_axis_data, z_axis_data, "Cosine of X graph", "X", "Cos(x)")
```

```python
def generate_data_2(range):
  x = np.random.randn(range)

  np.random.seed(40)
  y =  3 * x + 1

  np.random.seed(20)
  z = 8 * x

  return x, y, z



x_data, y_data, z_data = generate_data_2(10)
fig, axs = plt.subplots(1, 2, figsize = (10 ,4))
axs[0].scatter(y_data, x_data, s = 10, c = 'red', label = "Straight line 1")
axs[0].scatter(y_data,z_data,s = 10,c = 'blue',label = "Straight line 2")
axs[0].legend()
axs[0].set_xlabel("X data")
axs[0].set_ylabel("Y data")

# format strings
# fmt = [marker][line][color]
# for marker style - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
# marker style modified in line plot

axs[1].plot(y_data, x_data, '8-.m', label = "Straight line 1")
axs[1].plot(y_data,z_data,'4-.c',label = "Straight line 2")
axs[1].legend()
axs[1].set_xlabel("X data")
axs[1].set_ylabel("Y data")
"""
To modify location of legends in plots
==================   =============
    Location String      Location Code
    ==================   =============
    'best' (Axes only)   0
    'upper right'        1
    'upper left'         2
    'lower left'         3
    'lower right'        4
    'right'              5
    'center left'        6
    'center right'       7
    'lower center'       8
    'upper center'       9
    'center'             10
    ==================   =============
"""

plt.show()
```
